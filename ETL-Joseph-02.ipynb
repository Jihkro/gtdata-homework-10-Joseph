```python
#Import Dependencies
import pandas as pd
import numpy as np
import math
import IPython
```

```python
#load the data
#skip the first three rows as this contains just title information
arrest_df = pd.read_excel('table_69_arrest_by_state_2013.xls',skiprows=3)
```

```python
#drop last few rows as this contains just notes about the data
#See the specific notes here: https://ucr.fbi.gov/crime-in-the-u.s/2013/crime-in-the-u.s.-2013/tables/table-69/table_69_arrest_by_state_2013.xls/@@template-layout-view?override-view=data-declaration

arrest_df = arrest_df[0:102]
```

```python
#Clean the keys, replacing newline characters with spaces and removing numbers
cleanedkeys = []
for key in list(arrest_df.keys()):
    currentkey = key
    for y in range(10):
        currentkey = currentkey.replace(f'{y}','')
    currentkey = currentkey.replace('\n',' ')
    currentkey = str.strip(currentkey)
    cleanedkeys.append(currentkey)

arrest_df.rename(columns=dict(zip(list(arrest_df.keys()),cleanedkeys)),inplace=True)
```

```python
#Trim and replace statenames, removing spaces, commas, and numbers
#simultaneously duplicate missing fields for number of agencies and population total from row above

waitingmessages = [r' - ', r' \ ' , r' | ', r' / ']

for x in np.arange(1,102,2):
    IPython.display.clear_output(wait=True)
    print(f'{waitingmessages[int(((x+1)/2))%4]} : {x}')
    try:
        if math.isnan(arrest_df['State'][x]):
            arrest_df['State'][x] = arrest_df['State'][x-1]
    except:
        pass
    
    
    arrest_df['State'][x] = arrest_df['State'][x].replace(" ","")
    arrest_df['State'][x] = arrest_df['State'][x].replace(",","")
    for y in range(10):
        arrest_df['State'][x] = arrest_df['State'][x].replace(f"{y}","")
    
    if math.isnan(arrest_df['estimated  population'][x]):
        arrest_df['estimated  population'][x] = arrest_df['estimated  population'][x-1]
    if math.isnan(arrest_df['Number of agencies'][x]):
        arrest_df['Number of agencies'][x]=arrest_df['Number of agencies'][x-1]

    
    
```

```python
#Drop data for 18 and under, keeping only information for total population
arrest_df.drop(index=list(np.arange(0,102,2)),inplace=True)
#Drop data for District of Columbia
arrest_df.drop(index=17,inplace=True)
#Reset the index
arrest_df.reset_index(drop=True)
```
